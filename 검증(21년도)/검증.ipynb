```python
import pandas as pd
import joblib
from sklearn.metrics import roc_auc_score, accuracy_score
```

#### 최종 부도예측 모형에 사용할 데이터 형태는 재무비율 + 부도기사비율
#### 모델은 SVM

```python
상폐기업 = pd.read_csv('./datasets/상폐기업데이터.csv')
상폐기업['부도기업여부'] = 1
정상기업 = pd.read_csv('./datasets/정상기업데이터.csv')
정상기업['부도기업여부'] = 0

최종 = pd.concat([상폐기업, 정상기업])
최종.reset_index(drop=True, inplace=True)

input = 최종.iloc[:,2:8]
target = 최종.iloc[:,8]
```

```python
## 훈련한 모델과 scaler 로드
ml_model = joblib.load('./model/svc_model.pkl')
robscaler = joblib.load('./model/robustscaler.pkl')

## 독립변수들을 정규화하고 예측
input = robscaler.transform(input)

pred_proba = ml_model.predict_proba(input)
pred = ml_model.predict(input)

acc = accuracy_score(target, pred)
roc = roc_auc_score(target, pred_proba[:,1])

print(f'21년 데이터에 대한 accuracy : {acc}\n 21년 데이터에 대한 roc_auc_score : {roc}')
```

```python
최종['검증'] = pred
최종
```

```python
최종.to_csv('최종.csv', index=None, encoding='utf-8-sig')
```
