```python
import pandas as pd
```

```python
data=pd.read_csv("./data/최종기업목록.csv",encoding="utf-8-sig")
정상기업리스트=data[data["부실기업여부"]==0]
정상기업리스트_1=정상기업리스트.groupby(["회사명","회계년도"]).sum()
정상기업리스트_1=정상기업리스트_1.reset_index()
정상기업리스트_1
```

```python
부도기사비율=pd.read_csv("./정상기업_부도기사 비율.csv",encoding="utf-8-sig")
# 부도기사비율.columns=["회사명","부도기사 비율"]
부도기사비율.columns=["회사명","회계년도","부도기사 비율"]
# 부도기사비율['회계년도'] = 부도기사비율['회계년도'].astype('int64')
부도기사비율["회계년도"]=부도기사비율["회계년도"]-2
```

```python
df_OUTER_JOIN = pd.merge(정상기업리스트_1, 부도기사비율, on=['회사명','회계년도'], how='left')
# df_OUTER_JOIN.drop(["기업"],axis=1,inplace=True)
# df_OUTER_JOIN.fillna(0,inplace=True)
# df_OUTER_JOIN=df_OUTER_JOIN[["회사명","회계년도","폐지일자","운전자산총자본비율","이익잉여금총자산비율","총자산이익율","시장가부채비율","매출액회전율","부도기사 비율","부실기업여부"]]
# df_OUTER_JOIN.to_csv("정상기업_부도기사비율_재무비율.csv",encoding="utf-8-sig")
df_2=df_OUTER_JOIN[df_OUTER_JOIN["부도기사 비율"].isnull()==True]

df_2
# df_2.to_csv("정상기업_기사없는거.csv",encoding="utf-8-sig")
```
