폐지유사도 조합만들기

```python
폐지유사도단어리스트=pd.read_csv("상폐_폐지유사도단어리스트.csv",index_col=0)
단어리스트=list(폐지유사도단어리스트["단어"])
단어리스트
```

```python
from itertools import combinations

res = list(combinations(단어리스트, 3))
print(f"모든 경우: {res}")
print(f"모든 경우의 수: {len(res)}") 
```

```python
list_word=[]
for i in range(len(res)):
    list_word.append(list(res[i]))
    list_word[i].append("폐지")
    
    
len(list_word)
```

```python
list_word
```

기사에 들어갈 것들 비교

```python
list__기사수차이=[]
for i in range(len(list_word)):
    a="|".join(list_word[i])
    df_1=새롭게합친_상폐기사_df[새롭게합친_상폐기사_df['뉴스기사본문전처리'].str.contains(a) ]
    df_2=새롭게합친_정상기사_df[새롭게합친_정상기사_df['뉴스기사본문전처리'].str.contains(a) ]
    list__기사수차이.append(len(df_1)-len(df_2))
    
```

데이터프레임 만들기

```python
기사수조합비교_df=pd.DataFrame({'단어조합':list_word,'기사수 차이':list__기사수차이})
기사수조합비교_df.to_csv("mecab_fasttext_기사수조합비교.csv",encoding="utf-8-sig")
기사수조합비교_df["기사수 차이"].max() #1143
기사수조합비교_df[기사수조합비교_df["기사수 차이"]==1166]
```
