```python
상폐기업=pd.read_csv(r"C:\Users\jangd\OneDrive\사진\문서\GitHub\project2\data\상폐_새로합친거.csv",index_col=0)
상폐기업
```

```python
# 정상기업.reset_index(inplace=True)
# 정상기업.drop(["index"],axis=1,inplace=True)
# 정상기업

상폐기업.reset_index(inplace=True)
상폐기업.drop(["index"],axis=1,inplace=True)
상폐기업
```

```python
상폐기업["뉴스기사 레이블"]=상폐기업["뉴스기사본문전처리"]
```

```python
keywords=["상폐", "적격성", "이의신청", "폐지"]
for i in range(len(상폐기업)):
    if any(keyword in str(상폐기업["뉴스기사본문전처리"][i]) for keyword in keywords)==True:
        상폐기업['뉴스기사 레이블'][i]=1
    else:
        상폐기업['뉴스기사 레이블'][i]=0
```

```python
# 정상기업
상폐기업
```

```python
# 기사수=정상기업.groupby("기업").count()
기사수=상폐기업.groupby("기업").count()

기사수
```

```python
# 레이블합=정상기업.groupby("기업").sum()
레이블합=상폐기업.groupby("기업").sum()
레이블합
```

```python
df=기사수.reset_index()["기업"]
df=pd.DataFrame(df)
df["부도기사비율"]=df["기업"]
df
```

```python
for i in range(len(df)):
    df["부도기사비율"][i]=레이블합["뉴스기사 레이블"][i]/기사수["뉴스기사 레이블"][i]*100
df.to_csv("okt_상폐기업부도기사비율.csv",encoding="utf-8-sig")
```
