```python
import pandas as pd
```

### 데이터 불러오기

```python
df_재무비율=pd.read_csv('../Step1-1_재무비율전처리/최종기업목록_직접선정.csv', encoding='utf-8-sig')
```

```python
df_재무비율.dropna(inplace=True)
df_재무비율
```

### 정상기업

```python
df_정상기업_재무비율=df_재무비율[df_재무비율["부실기업여부"]==0]
df_정상기업_재무비율.reset_index(inplace=True)
df_정상기업_재무비율.drop("index",inplace=True,axis=1)
df_정상기업_재무비율
```

### 부도기업

```python
df_부실기업_재무비율=df_재무비율[df_재무비율["부실기업여부"]==1]
df_부실기업_재무비율.reset_index(inplace=True)
df_부실기업_재무비율.drop("index",inplace=True,axis=1)
df_부실기업_재무비율
```

#### 정상기업 부도기사비율

```python
df_정상기업_부도기사비율=pd.read_csv("./datasets/정상기업_부도기사 비율.csv",encoding="utf-8-sig")
df_정상기업_부도기사비율.columns=["회사명","회계년도","부도기사 비율"]
df_정상기업_부도기사비율["회계년도"]=df_정상기업_부도기사비율["회계년도"]-2 # 회계년도와 기사발행년도는 
df_정상기업_부도기사비율
```

### 부도기업 부도기사비율

```python
df_상폐기업_부도기사비율=pd.read_csv("./datasets/상폐기업_부도기사 비율.csv",encoding="utf-8-sig")
df_상폐기업_부도기사비율.columns=["회사명","부도기사 비율"]
df_상폐기업_부도기사비율
```

### 부도기업 부도기사비율합치기

```python
상폐레이블=pd.merge(left = df_부실기업_재무비율 , right = df_상폐기업_부도기사비율, how = "left", on = "회사명")
상폐레이블 = 상폐레이블[['회사명', '회계년도', '운전자산총자본비율', '이익잉여금총자산비율', '총자산이익율', '시장가부채비율', '매출액회전율',
       '폐지일자', '부도기사 비율', '부실기업여부']]
상폐레이블
```

### 정상기업 부도기사비율 합치기

```python
정상레이블=pd.merge(left = df_정상기업_재무비율 , right = df_정상기업_부도기사비율, how = "left", on = ["회사명","회계년도"])
정상레이블 = 정상레이블[['회사명', '회계년도', '운전자산총자본비율', '이익잉여금총자산비율', '총자산이익율', '시장가부채비율', '매출액회전율',
       '폐지일자', '부도기사 비율', '부실기업여부']]
정상레이블
```

### 데이터 총합 합치기

```python
data_final=pd.concat([상폐레이블,정상레이블],axis=0)
data_final.to_csv('../Step4-3_재무비율_워드임베딩_모델링/datasets/부도기사비율_재무비율.csv',encoding="utf-8-sig")
```
