```python
import pandas as pd
```

```python
data=pd.read_csv("./data/최종기업목록.csv",encoding="utf-8-sig")
부실기업리스트=data[data["부실기업여부"]==1]
부실기업리스트
```

```python
부도기사비율=pd.read_csv("./data/상폐기업_부도기사 비율.csv",encoding="utf-8-sig")
부도기사비율.columns=["회사명","부도기사 비율"]
```

```python
df_2=pd.merge(부실기업리스트, 부도기사비율, on='회사명', how='left')
df_2.isnull().sum()
# # df_2[df_2['부도기사 비율'].isnull()]
# df_2[df_2["회사명"]=="선팩테크"]["회사명"]=="유비트론"
# df_2[df_2["회사명"]=="유에이블"]="디웍스 글로벌"
# df_2[df_2["회사명"]=="유에이블"]["회사명"]="디웍스 글로벌"
# df_2[df_2["회사명"]=="프리젠"]["회사명"]="프리젠 오리엔트 프리젠"
# df_2[df_2["회사명"]=="금강제강"]["회사명"]="에이치디시에스"
# 프리젠 오린테프리젠트,스맥스
# 금강제강 = 에이치디시에스
# df_2["회사명"]
df_3=df_2[df_2['부도기사 비율'].isnull()]
df_3["회사명"]=["유비트론","디에이치패션","인스프리트","에이치디시에스","미주제강","디웍스 글로벌","지티앤티","오리엔트 프리젠"]
df_3
```

```python
df_OUTER_JOIN = pd.merge(부실기업리스트, 부도기사비율, left_on='회사명', right_on='기업', how='outer')
df_OUTER_JOIN.drop(["기업"],axis=1,inplace=True)
df_OUTER_JOIN.fillna(0,inplace=True)
df_OUTER_JOIN=df_OUTER_JOIN[["회사명","회계년도","폐지일자","운전자산총자본비율","이익잉여금총자산비율","총자산이익율","시장가부채비율","매출액회전율","부도기사 비율","부실기업여부"]]
df_OUTER_JOIN.to_csv("상폐기업_부도기사비율_재무비율.csv",encoding="utf-8-sig")
```

```python
df_3=df_3[["회사명","폐지일자"]]
df_3.reset_index(inplace=True)
df_3.drop(["index"],inplace=True,axis=1)
df_3.loc[8] = ['스맥스', '2017-02-24']
df_3
```
