```python
import pandas as pd
from eunjeon import Mecab
```

### 데이터 불러오기

```python
상폐기업뉴스 = pd.read_csv('../Step3-1_뉴스전처리/전처리후/상폐기업뉴스전처리.csv')
상폐기업뉴스추가 = pd.read_csv('../Step3-1_뉴스전처리/전처리후/상폐기업뉴스전처리추가.csv')
상폐기업 = pd.concat([상폐기업뉴스, 상폐기업뉴스추가])
```

```python
상폐기업.reset_index(inplace=True)
상폐기업.drop(["index"],axis=1,inplace=True)
상폐기업
```

### 전처리된 뉴스기사본문 토큰화

```python
mecab=Mecab()
## 형태소 분석기 mecab를 사용한 토큰화 작업 (다소 시간 소요)

## Mecab으로 명사만 추출
상폐기업['뉴스기사본문전처리'] = 상폐기업['뉴스기사본문전처리'].apply(lambda x: mecab.nouns(x))
## 한 글자 단어 제거
상폐기업['뉴스기사본문전처리'] = 상폐기업['뉴스기사본문전처리'].apply(lambda x: [word for word in x if len(word) > 1])
상폐기업
```

### 분류단어 넣기

```python
분류단어 = ['폐지', '격성', '퇴출', '실질', '심사', '속개', '이의', '심의', '거절']
분류단어 = '|'.join(분류단어)
분류단어
```

### 부도기사 레이블화

```python
라벨 = 상폐기업['뉴스기사본문전처리'].apply(lambda x: [word for word in x if word in 분류단어]) # 분류단어 들어간 것들
라벨 = 라벨.apply(lambda x: [word for word in x if word not in '상장']) # 상장이라는 단어만 있는 것을 제외
라벨 = 라벨.apply(lambda x: 1 if (len(x)>=1) else 0) # 한글자 제거
상폐기업['뉴스기사 레이블'] = 라벨
```

```python
상폐기업
```

### 기업 당 뉴스기사 수 추출

```python
기사수=상폐기업.groupby("기업").count()
기사수
```

### 기업 당 뉴스 부도뉴스 레이블 합 도출

```python
레이블합=상폐기업.groupby("기업").sum()
레이블합
```

### 부도기사비율 도출

```python
기사수["부도기사비율"]=기사수["뉴스기사 레이블"]
```

```python
기사수["부도기사비율"]=레이블합["뉴스기사 레이블"]/기사수["뉴스기사 레이블"]*100
```

```python
기사수
```

```python
기사수.reset_index(inplace=True)
기사수.drop(["기사발행일","기사제목","뉴스기사본문전처리","뉴스기사 레이블"],axis=1,inplace=True)
기사수.to_csv('./datasets/상폐기업_부도기사 비율.csv', index=None)
```
